```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
x_sample = df['T2'].to_numpy().reshape(-1,1)
y_sample = df['T6'].to_numpy().reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
x_sample_train, x_sample_test, y_sample_train, y_sample_test = train_test_split(x_sample, y_sample)
lr.fit(x_sample_train, y_sample_train)
y_sample_pred = lr.predict(x_sample_test)
print(r2_score(y_sample_test, y_sample_pred))
```

```python
df['date_col'] = pd.to_datetime(df['date'])
df = df.drop(columns=['date', 'lights'])
df.info()
```

```python
df['day'] = date_col.dt.day
df['month'] = date_col.dt.month
df['hour'] = date_col.dt.hour
df['minutes'] = date_col.dt.minute
df = df.drop(columns=['date_col'])
df.head()
```

```python
corr_matrix = df.corr()
corr_matrix['Appliances'].sort_values(ascending=False)
```

```python
y = df['Appliances']
X = df.drop(columns=['Appliances'])
X.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr2 = LinearRegression()
lr2.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
y_pred = lr2.predict(X_test)
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred, squared=False))
```

```python
#To find the residual sum of squares

rss = np.sum(np.square(y_test - y_pred))
print(rss)
```

```python
importance = lr2.coef_
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(X_train, y_train)
y_pred_ridge = ridge_reg.predict(X_test)
print(mean_squared_error(y_test, y_pred_ridge, squared=False))
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)
y_pred_lasso = lasso_reg.predict(X_test)
print(mean_squared_error(y_test, y_pred_lasso, squared=False))
```

```python
importance_lasso = lasso_reg.coef_
for i,v in enumerate(importance_lasso):
    print('Feature: %0d, Score: %.5f' % (i,v))
```
